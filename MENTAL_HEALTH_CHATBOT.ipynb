```python
import random
import matplotlib.pyplot as plt

name = input("Bot: Hello! What is your name? ")
print("Bot: Nice to meet you,", name)

mood_history = []

motivations = [
    "You are stronger than you think.",
    "Small progress is still progress.",
    "Believe in yourself and keep going.",
    "Every day is a fresh start.",
    "Your hard work will pay off.",
    "Stay positive and keep learning."
]
stress_words = ["stress", "stressed", "pressure", "overwhelmed"]
sad_words = ["sad", "upset", "unhappy", "depressed"]
happy_words = ["happy", "great", "good", "excited"]
tired_words = ["tired", "exhausted", "sleepy"]
lonely_words = ["lonely", "alone", "isolated"]
bored_words = ["bored", "boring", "nothing to do"]

while True:

    feeling = input(name + ": ").lower()

    if feeling == "bye":
        print("Bot: Take care of yourself. Have a good day!")
        break

    elif "motivate" in feeling or "motivation" in feeling:
        print("Bot:", random.choice(motivations))

    elif any(word in feeling for word in stress_words):
        responses = [
            "Stress can be tough. Try taking a short break.",
            "Maybe try deep breathing for a few minutes.",
            "Going for a short walk might help clear your mind."
        ]
        print("Bot:", random.choice(responses))
        mood_history.append(1)

    elif any(word in feeling for word in sad_words):
        print("Bot: I'm sorry you're feeling sad.")
        print("Bot: Talking to a friend can help.")
        mood_history.append(1)

    elif "anxious" in feeling:
        print("Bot: Anxiety can be tough.")
        print("Bot: Try the 4-7-8 breathing technique.")
        mood_history.append(1)

    elif any(word in feeling for word in tired_words):
        print("Bot: You might be overworking yourself.")
        print("Bot: Try getting some rest.")
        mood_history.append(1)

    elif any(word in feeling for word in happy_words):
        print("Bot: That's wonderful to hear! Keep doing what makes you happy.")
        mood_history.append(3)

    elif any(word in feeling for word in lonely_words):
        print("Bot: Feeling lonely is hard.")
        print("Bot: Try reaching out to someone you trust.")
        mood_history.append(2)

    elif any(word in feeling for word in bored_words):
        print("Bot: Feeling bored happens sometimes.")
        print("Bot: Maybe try music, dancing, or learning something new.")
        mood_history.append(2)

    else:
        print("Bot: Thank you for sharing. Remember you are not alone.")
        mood_history.append(2)





# Save mood history after chat ends
with open("mood_history.txt", "w") as file:
    file.write(str(mood_history))

print("Bot: Your mood data has been saved.")

# Show mood graph
plt.plot(mood_history, marker='o')
plt.title("Mood Tracker")
plt.xlabel("Conversation Step")
plt.ylabel("Mood Level")
plt.show()
```
